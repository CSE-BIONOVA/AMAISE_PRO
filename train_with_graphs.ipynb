```python
import matplotlib.pyplot as plt
# Open the text file for reading
with open('Result_9000.txt', 'r') as file:
    # Read lines from the file and store them in an array
    lines = file.readlines()

# Now 'lines' contains all the lines from the file as elements of an array
# You can access individual lines using array indexing
train_steps = (324000*0.8)//128
val_steps = (324000*0.2)//128
train_losses = []
val_losses = []
epochs = []
# Example: Print all lines
for i in range(len(lines)-1):
    train_loss = float(lines[i].strip().split(",")[1].split(":")[1])/train_steps
    train_losses.append(train_loss)
     # .strip() removes the newline character at the end of each line
    val_loss = float(lines[i].strip().split(",")[3].split(":")[1])/val_steps
    val_losses.append(val_loss)
    epochs.append(i+1)
fig, ax = plt.subplots(figsize=(10,7))
ax.plot(epochs, train_losses, label="train loss")
ax.plot(epochs, val_losses, label="val loss")
ax.set_xlabel("Epochs")
ax.set_ylabel("Loss")
ax.legend()
```
